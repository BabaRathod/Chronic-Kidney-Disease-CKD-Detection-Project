### According to the data description
- Cols(pcv, wc and rc) needs to convert back in numerical since it is object right now
- Cols(sg, al and su) should be nominal , convert from float to object

```python
# Categorical cols like specific_gravity, albumin and sugar which is float type right now
# Converting back to nominal data type categorical

# Convert float columns to categorical data type
categorical_columns = ['specific_gravity', 'albumin', 'sugar']
for col in categorical_columns:
    chronic_df[col] = chronic_df[col].astype('category')
```

```python
# Converting necessary columns like packed_cell_volume, white-blood_cell_count and red_blood_cell_count
# Currently it is in object type and converting back to numerical type

# Convert object columns to numerical data type
numerical_columns = ['packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count']
for col in numerical_columns:
    chronic_df[col] = pd.to_numeric(chronic_df[col], errors='coerce')  # Coerce invalid parsing to NaN
```

## **`Missing Value Treatment`**

```python
# Filling null values, we will use two methods, random sampling for higher null values and 
# Mean/mode sampling for lower null values

import random

# Creating function for imputing random values
def random_value_imputation(feature):
    random_sample = chronic_df[feature].dropna().sample(chronic_df[feature].isnull().sum(), random_state=0)
    random_sample.index = chronic_df[chronic_df[feature].isnull()].index
    chronic_df.loc[chronic_df[feature].isnull(), feature] = random_sample

# Creating function for imputing most common value (mode)
def impute_mode(feature):
    mode = chronic_df[feature].mode()[0]
    chronic_df[feature] = chronic_df[feature].fillna(mode)
```

```python
# Filling num_cols null values using random sampling method

# Apply random_value_imputation to fill null values in numerical columns
for feature in num_cols:
    if chronic_df[feature].isnull().sum() > 0:
        random_value_imputation(feature)
```

```python
# Let's check count of null values in num_cols again
chronic_df[num_cols].isnull().sum()
```

```python
# Filling "red_blood_cells" and "pus_cell" using random sampling method and rest of cat_cols using mode imputation

# Apply random_value_imputation to "red_blood_cells" and "pus_cell" columns
random_value_imputation("red_blood_cells")
random_value_imputation("pus_cell")

# Apply impute_mode for the rest of the categorical columns
for feature in cat_cols:
    if feature not in ["red_blood_cells", "pus_cell"]:
        impute_mode(feature)
```

## **`Model Building`**

```python
# Import KNeighborsClassifier, accuracy_score, confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn = KNeighborsClassifier(n_neighbors=5)  # Initialize KNN classifier with 5 neighbors
knn.fit(X_train, y_train)  # Train the classifier

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate training and test accuracy
train_acc = knn.score(X_train, y_train)
test_acc = knn.score(X_test, y_test)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Create classification report
class_report = classification_report(y_test, y_pred)

print(f"Training Accuracy of KNN is {train_acc}")
print(f"Test Accuracy of KNN is {test_acc}")
print("\nConfusion Matrix:\n")
print(conf_matrix)
print("\nClassification Report:\n")
print(class_report)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Create an instance of ExtraTreesClassifier
etc = ExtraTreesClassifier(random_state=42)

# Fit the classifier to the training data
etc.fit(X_train, y_train)

# Make predictions on the test data
y_pred_etc = etc.predict(X_test)

# Calculate training accuracy
etc_train_acc = etc.score(X_train, y_train)

# Calculate test accuracy
etc_test_acc = accuracy_score(y_test, y_pred_etc)

# Generate confusion matrix
conf_matrix_etc = confusion_matrix(y_test, y_pred_etc)

# Generate classification report
class_report_etc = classification_report(y_test, y_pred_etc)

# Print training and test accuracy, confusion matrix, and classification report
print("Extra Trees Classifier - Training Accuracy:", etc_train_acc)
print("Extra Trees Classifier - Test Accuracy:", etc_test_acc)
print("\nConfusion Matrix:\n", conf_matrix_etc)
print("\nClassification Report:\n", class_report_etc)
```
